#### Answer:
Vectors in math and physics are presenting shift, they can be described in math notation by arrow over variable like this $\vec A$\, they can be described in every dimension and also for each dimension they have 1 component
their most common properties are:
* Direction
* Shift
* Components
* Length - $\vec |A| = \sqrt (A_x^2 + A_y^2) $

Possible operations on vectors:\
* Scalar operations (addition, multiplication, division, subtraction) like $\vec A * 2 = (A_x * 2, A_y * 2)$ are simple operation that do operation on all components of vector (2 components on 2D vector, 3 components on 3D vector, … etc.)
* Non-scalar operations
    * Simple non-scalar operations (addition & subtraction) $ \vec A + \vec B = (A_x + B_x, A_y + B_y)$
    * dot product $ \vec A \cdotp \vec B = A_x * B_x + A_y * B_y $
    * vector product/cross product $ \vec A \times \vec B = \vec |A| * \vec |B| * \sin \theta $
